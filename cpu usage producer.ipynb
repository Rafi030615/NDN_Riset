```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_cpu_usage(cpu_data, value_data):
    plt.figure(figsize=(10, 5))
    plt.plot(cpu_data, linestyle='-', label="Ubuntu")
    plt.plot(value_data, linestyle='-', label= "OpenWRT")
    plt.xlabel("Index")
    plt.ylabel("CPU used")
    plt.title("Line Plot of CPU Usage")
    plt.grid(True)
    plt.legend()
    plt.show()
```
