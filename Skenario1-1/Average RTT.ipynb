# TEMPERATURE

```python
dataframes = []

for i in range(1, 11):
    filepath = rf"C:\Users\Lenovo\NDN_New\Skenario1-1\Data\Consumer\skenario1-1-{i}\cpu1-1-{i}\thermal-thermal_zone0\temperature-2024-02-19.csv"
    df = pd.read_csv(filepath)
    globals()[f't{i}'] = df
    dataframes.append(df)
    
for df in dataframes:
    df['epoch'] = pd.to_datetime(df['epoch'], unit='s')
    df['WIB'] = df['epoch'].dt.tz_localize('UTC').dt.tz_convert('Asia/Jakarta')
```
