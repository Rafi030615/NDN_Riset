```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
for i in range(1, 6):
    filepath = rf"Data\skenario1\log1-1-{i}.csv"
    globals()[f'z11{i}'] = pd.read_csv(filepath, index_col = False)
```

```python
dfs = [z111, z112, z113, z114, z115]
tinggi = []
for df in dfs:
    tinggi.append(df.loc[0]["AverageRTT(ms)"])

datar = [1, 2, 3,4, 5]
plt.figure(figsize=(10, 5))
plt.bar(datar, tinggi, width=0.5)

for i, t in enumerate(tinggi):
    plt.text(datar[i], t, str(t), ha="center", va="bottom")

plt.ylabel("Average RTT")
plt.title("RTT")
plt.show()
```
