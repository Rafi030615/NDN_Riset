```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframes11=[]

for i in range(1, 6):
    filepath = rf"Consumer\skenario1\skenario1-1-{i}\cpu1-1-{i}\memory\memory-used-2024-03-04.csv"
    df = pd.read_csv(filepath, index_col = False)
    globals()[f't11{i}'] = df
    dataframes11.append(df)
    
for df in dataframes11:
    df['epoch'] = pd.to_datetime(df['epoch'], unit='s')
    df['WIB'] = df['epoch'].dt.tz_localize('UTC').dt.tz_convert('Asia/Jakarta')
    
plt.figure(figsize=(20,10))
for i, df in enumerate(dataframes11, 1):
    plt.plot(df['WIB'], df['value'], label=f't11{i}') 
```
