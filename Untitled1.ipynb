```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframes33=[]

for i in range(1, 6):
    filepath = rf"Consumer\skenario3\skenario3-3-{i}\cpu3-3-{i}\memory\memory-used-2024-03-05.csv"
    df = pd.read_csv(filepath, index_col = False)
    globals()[f't33{i}'] = df
    dataframes33.append(df)
    
for df in dataframes33:
    df['epoch'] = pd.to_datetime(df['epoch'], unit='s')
    df['WIB'] = df['epoch'].dt.tz_localize('UTC').dt.tz_convert('Asia/Jakarta')
    
plt.figure(figsize=(20,10))
for i, df in enumerate(dataframes33, 1):
    plt.plot(df['WIB'], df['value'], label=f't33{i}') 
```
