```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
z1 = pd.read_csv("Data\Consumer\skenario1-1-1\log1-1-1.csv", index_col = False)
z2 = pd.read_csv("Data\Consumer\skenario1-1-2\log1-1-2.csv", index_col = False)
z3 = pd.read_csv("Data\Consumer\skenario1-1-3\log1-1-3.csv", index_col = False)
z4 = pd.read_csv("Data\Consumer\skenario1-1-4\log1-1-4.csv", index_col = False)
z5 = pd.read_csv("Data\Consumer\skenario1-1-5\log1-1-5.csv", index_col = False)
```

```python
z1 = z1.drop(z1.index[0])
z2 = z2.drop(z2.index[0])
z3 = z3.drop(z3.index[0])
z4 = z4.drop(z4.index[0])
z5 = z5.drop(z5.index[0])
```

```python
z1
```

```python
dfs = [z1, z2, z3, z4, z5]


for i, df in enumerate(dfs, 1):
    plt.figure(figsize=(20,20))
    plt.plot(df['PatternID'], df['InterestSent'], label='Interest')
    plt.ylabel('InterestSent')
    plt.xlabel('PatternID')
    plt.title(f'Zipf for z{i}')
    plt.show()
```
